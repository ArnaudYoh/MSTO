```python
N = 8192
p = 32
b = 0.001
a = 1

s_Re = [np.random.normal() * 1 / (b + 4 * a * np.sin(np.pi * p * k / N)**2) for k in range(N)]
s_Im = [np.random.normal() * 1 / (b + 4 * a * np.sin(np.pi * p * k / N)**2) for k in range(N)]
   
s = [complex(*x) for x in zip(s_Re, s_Im)]
s = np.fft.ifft(s)
s = s.real
plt.plot(s)
plt.show()
sd.play(s)
```

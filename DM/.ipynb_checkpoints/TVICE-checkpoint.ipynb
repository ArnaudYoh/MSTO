# Answers to the PDF questions

### Question 1: Loi marginale conditionnelle $\pi(u(i)|u(i^c))$

Nous avons l'égalité suivante:
$$
\begin{equation}
\pi(u(i)|u(i^c)) = p(u(i)|u(i^c), u_0(i))
\end{equation}
$$
A partir de cette égalité nous pouvons utiliser la définition des probabilité conditionnelle pour obtenir:
$$
\begin{equation}
p(u(i)|u(i^c), u_0(i)) = \frac{p(u_0(i)|u(i^c),u(i) ) * p(u(i)|u(i^c))}{p(u_0(i)|u(i^c))}
\end{equation}
$$

Notons que nous nous intéressons à $u(i)$ donc nous avons la relation de proportionalité:

$$
\begin{equation}
 \frac{p(u_0(i)|u(i^c),u(i)) * p(u(i)|u(i^c))}{p(u_0(i)|u(i^c))} \propto p(u_0(i)|u(i^c),u(i)) * p(u(i)|u(i^c))
\end{equation}
$$

Divisons notre problème entre $p(u_0(i)|u(i^c),u(i))$ et $p(u(i)|u(i^c))$. 

Pour $p(u_0(i)|u(i^c),u(i))$, nous remarquons que $p(u_0(i)|u(i^c),u(i)) = p(u_0(i)|u)$ ainsi

$$
\begin{equation}
p(u_0(i)|u(i^c),u(i)) = e^{- \frac{1}{2\sigma^2} |u(i) - u_0(i)|^2}
\end{equation}
$$ 

Pour $p(u(i)|u(i^c))$ nous pouvons utiliser le théorème de Bayes pour obtenir 

$$
\begin{align}
p(u(i)|u(i^c)) &= \frac{p(u(i), u(i^c))}{p(u(i^c))} \\
&= e^{- \frac{1}{2} \lambda \sum_{i\in\Omega, j\in N(i)} |u(i) - u(j)| - \left[ - \frac{1}{2} \lambda \sum_{i\in i^c, j\in N(i)} |u(i) - u(j)| + \frac{1}{2} \lambda \sum_{i\in i^c, j\in N(i) \setminus  i^c} |u(i) - u(j)| \right]} \\
&= e^{- \lambda \sum_{j\in N(i)} |u(i) - u(j)|}
\end{align}
$$
 
Ainsi, nous avons:
$$
\begin{align}
\pi(u(i)|u(i^c)) &= p(u(i)|u(i^c), u_0(i)) \\
&= e^{- \frac{1}{2\sigma^2} |u(i) - u_0(i)|^2} * e^{- \lambda \sum_{j\in N(i)} |u(i) - u(j)|} \quad \blacksquare
\end{align} 
$$

### Question 2: Espérance $\mathbb{E}_{U \sim \pi}[U(i)|U(i^c) = w(i^c)]$

Nous avons:
$$
\begin{equation}
\mathbb{E}_{U \sim \pi}[U(i)|U(i^c) = w(i^c)] = \int_\mathbb{R} u(i) \pi(u(i) | w(i^c)) d u(i)
\end{equation}
$$


Rappelons que $u(i) = u_0(i) + s$, ce qui nous donne:
$$
\begin{equation}
\int_\mathbb{R} u(i) \pi\left[u(i) | w(i^c)\right] d u(i) = \int_\mathbb{R} u_0(i)\pi\left[u(i)| w(i^c)\right] d u_0(i) + \int_\mathbb{R} s\pi\left[u(i) | w(i^c)\right] ds 
\end{equation}
$$


Commençons par simplifier $\int_\mathbb{R} u_0(i)\pi\left[u(i)| w(i^c)\right] d u_0(i)$. Nous omettrons les $du_0(i)$ pour plus de lisibilité:
$$
\begin{align}
\int_\mathbb{R} u_0(i)\pi\left[u(i)| w(i^c)\right] &= 
\int_\mathbb{R} u_0(i) \left[p(u_0(i)| u_0(i), w(i^c)) + p(s| u_0(i), w(i^c)) \right] \\
&=  u_0(i) + \int_\mathbb{R} u_0(i) p(s| u_0(i), w(i^c)) \\
&= u_0(i) + 0 \\
&= u_0(i)
\end{align}
$$

Pour $\int_\mathbb{R} s\pi\left[u(i) | u(i^c)\right] ds$ il suffit d'utiliser le résultat de la question précédente. Nous omettrons les $ds$ pour plus de lisibilité:
$$
\begin{align}
\int_\mathbb{R} s\pi\left[u(i) | u(i^c)\right] &= \int_\mathbb{R} s \frac{1}{p(u_0(i) | u(i^c))} e^{- \frac{1}{2\sigma^2} s^2} * e^{- \lambda \sum_{j\in N(i)} |u_0(i) + s - w(j)|} \\
&= \frac{1}{Z} \int_\mathbb{R} s e^{- \frac{1}{2\sigma^2} s^2} * e^{- \lambda \sum_{j\in N(i)} |u_0(i) + s - w(j)|}
\end{align}
$$
avec $Z = p(u_0(i) | u(i^c))$. 


Ainsi nous avons:
$$
\begin{equation}
\mathbb{E}_{U \sim \pi}[U(i)|U(i^c) = w(i^c)] = u_0(i) + \frac{1}{Z} \int_\mathbb{R} s e^{- \frac{1}{2\sigma^2} s^2} * e^{- \lambda \sum_{j\in N(i)} |u_0(i) + s - w(j)|}
\end{equation}
$$
avec $Z = p(u_0(i) | u(i^c))$. $\quad \blacksquare$
